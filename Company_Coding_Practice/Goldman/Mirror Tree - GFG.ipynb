```python
class Node:
    def __init__(self,key):
        self.data = key
        self.left = None
        self.right = None
a = Node(1)
b = Node(2)
c = Node(3)
a.left = c
a.right = b
d = Node(4)
e = Node(5)
b.left = e
b.right = d
```

```python
def printInorder(root):
    if root == None:
        return
    print(root.data)
    printInorder(root.left)
    printInorder(root.right)
```

```python
printInorder(a)
```

```python
def mirror(root):
    if root == None:
        return
    tmp = root.left
    root.left = root.right
    root.right = tmp
    mirror(root.left)
    mirror(root.right)
    return root
```

```python
printInorder(mirror(a))
```
