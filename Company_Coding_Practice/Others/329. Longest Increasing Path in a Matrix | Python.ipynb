Given an m x n integers matrix, return the length of the longest increasing path in matrix.
<br>
From each cell, you can either move in four directions: left, right, up, or down. You may not move diagonally or move outside the boundary (i.e., wrap-around is not allowed).
<br>


```python
matrix = [[9,9,4],[6,6,8],[2,1,1]]
```

```python
memo = {}
directions = [(0,1),(1,0),(-1,0),(0,-1)]
def dfs(row,col,matix):
    if (row,col) in memo:
        return memo[(row,col)]
    memo[(row,col)] = 1
    
    for x,y in directions:
        new_row = row + x
        new_col = col + y
        
        if 0 <= new_row < len(matrix) and 0<= new_col < len(matrix[0]) and matrix[row][col] < matrix[new_row][new_col]:
            memo[(row,col)] = max(memo[(row,col)], 1 + dfs(new_row,new_col,matrix))
    return memo[(row,col)]
    
def longestIncreasingPath(matrix):
    res = 1
    
    for row in range(len(matrix)):
        for col in range(len(matrix)):
            res = max(res,dfs(row,col,matrix))
    return res
```
