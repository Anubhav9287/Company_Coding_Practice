```python
class Node:
    def __init__(self,key):
        self.val = key
        self.left = None
        self.right = None
```

```python
a1 = Node(4)
a2 = Node(2)
a1.left = a2
a3 = Node(7)
a1.right = a3
a4 = Node(1)
a5 = Node(3)
a3.left = a4
a3.right = a5
```

```python
def parsing(root,result):
    if not root:
        return result
    result.append(root.val)
    return parsing(root.left,result[:])
    return parsing(root.right,result[:])

def search(root,target):
    if not root:
        return []
    if target < root.val:
        return search(root.left,target)
    elif target > root.val:
        return search(root.right,target)
    else:
        return root
```

```python
search(a1,2)
# parsing(a1,[])
```
