Given a binary tree, find the lowest common ancestor (LCA) of two given nodes in the tree.
<br>
According to the definition of LCA on Wikipedia: “The lowest common ancestor is defined between two nodes p and q as the lowest node in T that has both p and q as descendants (where we allow a node to be a descendant of itself).”
<br>


```python
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
a = TreeNode(3)
b = TreeNode(5)
c = TreeNode(1)
d = TreeNode(6)
e = TreeNode(2)
f = TreeNode(0)
g = TreeNode(8)
h = TreeNode(7)
i = TreeNode(4)
a.left = b
a.right = c
b.left = d
b.right = e
c.left = f
c.right = g
e.left = h
e.right = i
```

```python
def printF(root):
    if not root:
        return
    print(root.val)
    printF(root.left)
    printF(root.right)
printF(a)
```

```python
def lowestCommonAscestor(root,p,q):
    if root == None:
        return None
    if root == p or root == q:
        return root.val
    l = lowestCommonAscestor(root.left,p,q)
    r = lowestCommonAscestor(root.right,p,q)
    
    if l != None and r != None:
        return root.val
    else:
        return l or r
```

```python
p = 6
q = 2
print(lowestCommonAscestor(a,p,q))
```
